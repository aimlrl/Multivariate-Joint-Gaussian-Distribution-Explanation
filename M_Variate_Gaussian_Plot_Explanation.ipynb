```python
from matplotlib import cm

from mpl_toolkits.mplot3d import Axes3D
```

```python
import scipy.stats as s

import numpy as np
```

Let's consider two continous normally random variables X and Y having means of respectively 0 and 1 as well as standard deviations of 1 and 1.5. Let's sample 200 values from their population distribution. 

```python
X = np.linspace(-3,3,200)

Y = np.linspace(-3,4,200)
```

```python
X,Y = np.meshgrid(X,Y)
```

```python
data = np.empty(X.shape+(2,))
```

```python
data.shape
```

```python
data[:,:,0] = X

data[:,:,1] = Y
```

```python
mean_vec = np.array([0,1])
```

```python
cov_mat = np.array([[1,-0.5],[-0.5,1.5]])
```

```python
P_X_equals_x_and_Y_equals_y = s.multivariate_normal.pdf(data,mean=mean_vec,cov=cov_mat)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12,12))

axes = fig.gca(projection='3d')

axes.plot_surface(X=X,Y=Y,Z=P_X_equals_x_and_Y_equals_y,rstride=3,
                  cstride=3,linewidth=1,antialiased=True,cmap=cm.viridis)

axes.contourf(X=X,Y=Y,Z=P_X_equals_x_and_Y_equals_y,zdir='z',cmap=cm.viridis,offset=-0.15)

axes.set_zlim(-0.15,0.2)

axes.set_zticks(np.linspace(0,0.2,5))

axes.view_init(27,-21)

plt.show()
```
